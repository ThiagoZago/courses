**Exercício 1: Implemente as duas classes abaixo **

![title](img/diagrama_classe_disciplina.png)

```python
class Disciplina:
    def __init__(self, codigo, nome):
        self.__codigo = codigo
        self.__nome = nome
        
    def setNome(self, nome):
        self.__nome = nome
        
    def getNome(self):
        return self.__nome
    
    def setCodigo(self, codigo):
        self.__codigo = codigo
        
    def getCodigo(self):
        return self.__codigo
    
    
class Curso:
    def __init__(self, nomeCurso):
        self.__nomeCurso = nomeCurso
        self.disciplinas = []
        
    def criarDisciplina(self, codigo, nome):
        self.disciplinas.append(Disciplina(codigo, nome))
        
    def removerDisciplina(self, codigo):
        for i in range(len(self.disciplinas)):
            if self.disciplinas[i].getCodigo() == codigo:
                del(self.disciplinas[i])
                break
    
jogos = Curso("Jogos Digitais")    
jogos.criarDisciplina(0, "POO")
jogos.criarDisciplina(1, "Estruturas")

for disc in jogos.disciplinas:
    print(disc.getNome())
    
jogos.removerDisciplina(0)

for disc in jogos.disciplinas:
    print(disc.getNome())
```

```python
class Cliente:
    def __init__(self, codigo, nome, percentualDesconto):
        self.__codigo = codigo
        self.__nome = nome
        self.__percentualDesconto = percentualDesconto

    def setNome(self, nome):
        self.__nome = nome
        
    def getNome(self):
        return self.__nome
    
    def setCodigo(self, codigo):
        self.__codigo = codigo
        
    def getCodigo(self):
        return self.__codigo      

    def setPercentualDesconto(self, __percentualDesconto):
        self.__percentualDesconto = __percentualDesconto
        
    def getPercentualDesconto(self):
        return self.__percentualDesconto        
    
class NotaFiscal:
    def __init__(self):
        self.__numero = 0
        self.__valorTotal = 0
        self.__cliente = None    
            
    def setNumero(self, numero):
        self.__numero = numero
        
    def getNumero(self):
        return self.__numero
    
    def setValorTotal(self, valorTotal):
        self.__valorTotal = valorTotal
        
    def getValorTotal(self):
        return self.__valorTotal
    
    def getCliente(self):
        return self.__cliente
    
    def setCliente(self, cliente):
        self.__cliente = cliente
        
    def getValorAposDesconto(self):
        perc = 1-(self.getCliente().getPercentualDesconto()/100)
        return self.getValorTotal()*perc
            
cliente = Cliente(0, "FX", 20)        
nf = NotaFiscal()
nf.setNumero(1)
nf.setValorTotal(200)
nf.setCliente(cliente)
print(nf.getValorAposDesconto())
            
```

** Exercício 3: Implemente uma classe chamada Caixa, que serve para cadastrar clientes e notas fiscais. Sua classe deverá pedir que o usuário digite os dados do Cliente e da Nota Fiscal, evitando que aconteça duplicidades nos códigos**

![title](img/diagrama_classe_nf_cliente_caixa.png)

```python
class Cliente:
    def __init__(self, codigo, nome, percentualDesconto):
        self.__codigo = codigo
        self.__nome = nome
        self.__percentualDesconto = percentualDesconto

    def setNome(self, nome):
        self.__nome = nome
        
    def getNome(self):
        return self.__nome
    
    def setCodigo(self, codigo):
        self.__codigo = codigo
        
    def getCodigo(self):
        return self.__codigo      

    def setPercentualDesconto(self, __percentualDesconto):
        self.__percentualDesconto = __percentualDesconto
        
    def getPercentualDesconto(self):
        return self.__percentualDesconto        
    
class NotaFiscal:
    def __init__(self):
        self.__numero = 0
        self.__valorTotal = 0
        self.__cliente = None    
            
    def setNumero(self, numero):
        self.__numero = numero
        
    def getNumero(self):
        return self.__numero
    
    def setValorTotal(self, valorTotal):
        self.__valorTotal = valorTotal
        
    def getValorTotal(self):
        return self.__valorTotal
    
    def getCliente(self):
        return self.__cliente
    
    def setCliente(self, cliente):
        self.__cliente = cliente
        
    def getValorAposDesconto(self):
        perc = 1-(self.getCliente().getPercentualDesconto()/100)
        return self.getValorTotal()*perc
            
class Caixa:
    def __init__(self):
        self.__notas = []
        self.__clientes = []
        
    def criarNota(self):
        while True:
            numero = int(input("Digite o número da nota: "))
            if self.getNota(numero)!=None:
                print("Esse número já existe. Digite outro")
            else:
                valor = float(input("Digite o valor total: "))
                cliente = self.criarCliente()
                nf = NotaFiscal()
                nf.setNumero(numero)
                nf.setValorTotal(valor)
                nf.setCliente(cliente)
                self.__notas.append(nf)
                break
                
    def iniciarCaixa(self):
        while True:
            op = input("Escolha: 1) Criar cliente; 2) Inserir nota; 3) Sair ")
            if op=="3":
                print("Fechando o caixa...")
                break
            else:
                if op=="1":
                    self.criarCliente()
                elif op=="2":
                    self.criarNota()
                else:
                    print("Opção inválida")
                
    def criarCliente(self):
         while True:
            codigo = int(input("Digite o código do cliente: "))
            cliente = self.getCliente(codigo)
            if cliente!=None:
                op = input("Esse código de cliente já existe. Deseja usar esse cliente (s) ou criar um novo (n)? ")
                if op=="s":
                    self.__clientes.append(cliente)
                    return cliente
                else:
                    continue
            else:
                nome = input("Digite o nome do cliente: ")
                perc = float(input("Digite o percentual de desconto: "))
                cliente = Cliente(codigo, nome, perc)
                self.__clientes.append(cliente)
                return cliente
                
    def getNota(self, numero):
        for nota in self.__notas:
            if nota.getNumero()==numero:
                return nota
        return None

    def getCliente(self, codigo):
        for cliente in self.__clientes:
            if cliente.getCodigo()==codigo:
                return cliente
        return None
        
cx = Caixa()
cx.iniciarCaixa()
            
```
